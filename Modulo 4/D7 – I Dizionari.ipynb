```python
# 1. Carichiamo il dataset di amazon tramite pandas
import pandas as pd
amazon_data = pd.read_csv('beginner_datasets/amazon.csv')
amazon_data
```

```python
# Valutiamo le dimensioni del dataset e visualizziamo dieci righe a caso
print("Il dataset è composto da",amazon_data.shape[0],"righe e",amazon_data.shape[1],"colonne.")
amazon_data.sample(10)
```

```python
# Osserviamo quali sono i nomi di colonna
print("I nomi di colonna sono:")
print(amazon_data.columns[0])
print(amazon_data.columns[1])
```

```python
# Il dataset è bilanciato? Ovvero, il numero di recensioni positive è uguale a quello delle negative?
print("Le recensioni positive sono",amazon_data.Positive.value_counts()[1], "mentre quelle negative sono",amazon_data.Positive.value_counts()[0],".\nSi evice che il dataset non è bilanciato in favore delle recensioni positive.")
```

```python
# Quante recensioni parlano di app?
# Quante parlano di giochi?
app = amazon_data['reviewText'].str.contains('app').sum()
giochi = amazon_data['reviewText'].str.contains('game').sum()
print("Ci sono",app,"recensioni che parlano di app e",giochi,"che parlano di giochi")
```

```python
# In generale quante recensioni raccomandano il prodotto?
# E quante non lo raccomandano
racc_ = amazon_data['reviewText'].str.contains('recommend')
pos_ = amazon_data["Positive"] == True
neg_ = amazon_data["Positive"] == False
racc_pos = amazon_data.loc[racc_ & pos_,"reviewText"]
racc_neg = amazon_data.loc[racc_ & neg_,"reviewText"]
print("In totale ci sono ",racc_pos.count(),"che raccomandano il prodotto recensito,\n",racc_neg.count()," invece non lo raccomandano.")
# Stampate qualche esempio dell'una e dell'altra
print("\nDi seguito qualche esempio:\n \nPositive:")
print(racc_pos.sample(5))
print("\n\nNegative:")
print(racc_neg.sample(5))
```

```python
# Le recensioni che menzionano il Kindle sono perlopiù positive o negative?
Kindle_up = amazon_data['reviewText'].str.upper().str.contains("KINDLE")
kindle = amazon_data.loc[Kindle_up]
print("Le recensioni che menzionano il kindle sono perlopiù positive.")
print("Positive:",kindle["Positive"].value_counts()[1])
print("Negative:",kindle["Positive"].value_counts()[0])
```

```python
# Osserviamone le dimensioni e un'anteprima di cinque righe
import pandas as pd
diab = pd.read_csv("beginner_datasets/diabetes.csv")
print("Colonne:",diab.shape[1],"\nRighe:",diab.shape[0])
diab.head(5)
```

```python
# Prendiamoci un pò di tempo per dare un'occhiata ai metadati delle colonne
print("I Metatdati:\n"+"-"+"\n-".join(diab.columns))
#Stampiamo i descrittori statistici del dataset
diab.describe()
```

```python
# Selezioniamo i dati relativi a diverse fasce di età : <20,20-30,30-40,40-50,>50
fascia_1 = diab['Age (years)'] < 20
fascia_2 = diab['Age (years)'].between(20,30)
fascia_3 = diab['Age (years)'].between(30,40)
fascia_4 = diab['Age (years)'].between(40,50)
fascia_5 = diab['Age (years)'] > 50
# Qual 'è la media della pressione sanguigna diastolica per le diverse fasce di età?
print("La media della pressione sanguigna è:\n"+
      str(round(diab[fascia_1]['Diastolic blood pressure (mm Hg)'].mean(),2))+"\n"+
      str(round(diab[fascia_2]['Diastolic blood pressure (mm Hg)'].mean(),2))+"\n"+
      str(round(diab[fascia_3]['Diastolic blood pressure (mm Hg)'].mean(),2))+"\n"+
      str(round(diab[fascia_4]['Diastolic blood pressure (mm Hg)'].mean(),2))+"\n"+
      str(round(diab[fascia_5]['Diastolic blood pressure (mm Hg)'].mean(),2)))
# Qual è la media della pressione per ogni singolo anno di età?
pressione_anno = diab.groupby(by="Age (years)")['Diastolic blood pressure (mm Hg)'].aggregate(['mean'])
pressione_anno
```

```python
# carichiamo il dataset insurance.csv
import pandas as pd
insu_ = pd.read_csv('beginner_datasets/insurance.csv')
# visualizziamo le dimensioni, un'anteprima e oserviamo i nomi di colonna
print("Colonne:",insu_.shape[1],"\nRighe:",insu_.shape[0],
      "\nI nomi delle colonne sono:\n-"+"\n-".join(insu_.columns))
insu_.sample(8)
```

```python
# Quali sono le medie di charges rispetto a region, smoker e sex?
reg_=insu_.groupby('region')['charges'].aggregate('mean')
smo_=insu_.groupby('smoker')['charges'].aggregate('mean')
sex_=insu_.groupby('sex')['charges'].aggregate('mean')
print(reg_)
print(sex_)
print(smo_)
("La differenza sostianziale è fra fumatori e non")
```

```python
# Quali sono i descrittori statistici di bmi? Quali sono minimo, media e massimo di charges rispetto ai diversi quartili dei valori di bmi
bmi_ = insu_.groupby("bmi",)['charges'].aggregate(['max','min','mean'])
print(insu_['bmi'].describe())
print(bmi_)
```

```python
# scarichiamo il dataset Pokemon
pkm = pd.read_csv('beginner_datasets/pokemon.csv')
# visualizziamo le dimensioni, un'anteprima e oserviamo i nomi di 
print("Colonne:",pkm.shape[1],"\nRighe:",pkm.shape[0],
      "\nI nomi delle colonne sono:\n-"+"\n-".join(pkm.columns))
pkm.sample(8)
```

```python
#settare la prima colonna ocme indice
pkm = pkm.set_index("#")
```

```python
# Ci sono ripetizioni, ovvero righe che si riferiscono allo stesso pokemon?
pkm.drop_duplicates()
```

```python
# Per ogni tipo in Type 1, quantio ce ne sono?
pkm.groupby('Type 1')['Type 1'].count()
```

```python
# Quali sono i pokemeno legendari
legend_ = pkm['Legendary'] == True
pkm[legend_]
```

```python
# E quali sono Leggendari e di Type 1 Grass?
grass_ = pkm['Type 1'] == 'Grass'
pkm[legend_ & grass_]
```

```python
# E i leggendari di tipo 1 Ice o Fire?
ice_ = pkm['Type 1'] == "Ice"
fire_ = pkm['Type 1'] == "Fire"
tp_1 = ice_ | fire_
pkm[legend_ & tp_1]
```

```python
# Ordiniamo il dataset per la colonna name
pkm.sort_values("Name")
pkm.set_index("Name")
```

```python
# Come accediamo alle statistiche di tutti i pokemon il cui nome inizia per D
start_D = pkm["Name"].str.startswith('D')
pkm[start_D]
```

```python
# Quali sono i pokemon della prima generazione con attacco >50 e HP <60
gen_1 = pkm['Generation'] == 1
attack = pkm['Attack'] > 50
HP = pkm['HP'] < 60
pkm.loc[gen_1 & attack & HP]
# Non ci sono dati nulli
```
